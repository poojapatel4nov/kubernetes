# Matplotlib.pyplot
creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.
In matplotlib.pyplot various states are preserved across function calls, so that it keeps track of things like the current figure and plotting area, and the plotting functions are directed to the current axes (please note that "axes" here and in most places in the documentation refers to the axes part of a figure and not the strict mathematical term for more than one axis).



```python
import matplotlib.pyplot as plt
import numpy as np

a = np.array([1,2,3,4,6,7,8,9,10])
b = np.array([1,2,3,4,6,7,8,9,10])
```

```python
plt.title("Linear Line")
plt.xlabel("a")
plt.ylabel("b")
plt.plot(a,b)
plt.show()
```

```python
g = np.array([3,6,44,12,67,45,55])
h = np.array([22,4,5,6,78,22,12])
```

```python
plt.xlabel("g")
plt.ylabel("h")
plt.plot(g,h)
plt.show()
```

```python
plt.plot(g,h, '--g')
plt.show()
```

```python
#Histogram

#A histogram take data in a series and devide the data into a number of bins, it then print frequency of data into each bins(i.e. interval of points), it is useful in understanding the cound of data range
#we should used this when we need the count of the vaiable in a plot


a= np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins= [0,20,40,60,80,100])

plt.title("historgam")
plt.show()
```
