```python
#ChiSquare test check differenece or relationship between two categorical data.

import pandas as pd
import numpy as np
import scipy.stats as stats

ds = pd.read_csv("role_website.csv")
```

```python
df = pd.DataFrame(ds)
df
```

# Chi Squar Test
by default chi square test has two hypothesis
H0--> Two catagorical data are independent(Null Hypothesis)
Ha --> Two Categorical data are dependent(Altenate Hypothesis)


```python
df.describe()
```

```python
df.info()
```

```python
df.dtypes
```

```python
#Create Contigency Table

df_crosstab = pd.crosstab(df['role'],df['website'],margins=True)

df_crosstab
```

```python
#To Calculate obsarved data for ch square test , we genrate Contigency table and there index(student, teacher ) column(facebook and linkdin) is obsarved data.

df_crosstab.columns=["facebook","linkedin","row_totals"]
df_crosstab.index=["student","teacher","col_totals"]

df_crosstab
```

```python
#Get the obsarved table


observed = df_crosstab.iloc[0:2,0:2]
observed
```

```python
#How expected values are calculated internally
#E = (row_Total*column_total)/grand_total
#E1 = 13*16/34
#E2= 13*18/34
#E3 =21*16/34
#E4 = 21*18/34

expected = np.outer(df_crosstab['row_totals'][0:2],df_crosstab.loc["col_totals"][0:2])/34
expected
```

```python
#Calculate ChiSquare using scipy.stat library
chi2,p,dof,expected=stats.chi2_contingency(observed=observed)
#We can check the returned data by calling it
#The returnd data includes chi_square_stat, pvalue, df, expected_crosstab
#print("stresult=", stresult)
print("Chi Square Test:",chi2)
print("p-values=",p)   
print("dof", dof)                #Degree of freedom
print("expected", expected)      #Same value which we calculated above using np.outer
```

From this we interpret that p value is .016 that is less then .05 --> significant value hence null hypothesis is rejected  and shows that two variable are dependent on each other.

```python
#Significance value
alpha = 0.5
#determine whether to reject or keep your null hypothesis
print("Significance level alpha=%.2f, p=%.2f" %(alpha,p))
if p >= alpha:
    print('Variables are not associated and independent(accept H0 - Null hypothesis)')
else:
    print('Variables are associated or dependent (Reject H0 and Accept H1)')
    
```
