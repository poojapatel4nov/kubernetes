# Seaborn Visualization Library 
Seaborn is built on Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
x = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
y =[4,7,5,3,6,5,4]
```

```python
ax = sns.stripplot(x,y)
ax.set(xlabel='Days', ylabel='Amount_spend')
plt.title("My First Graph")
plt.show()
```

```python
dsiris=pd.read_csv('iris.csv')
#print(dsiris)
print(dsiris.head())
sns.set(style="whitegrid")
ax = sns.stripplot(x='class',y ='sepal length', data=dsiris)
plt.title('Graph')
plt.show()
```

```python
sns.set(style="whitegrid")
ax = sns.stripplot(x='class',y ='petal length', data=dsiris)
plt.title('Graph')
plt.show()
```

```python
sns.set(style="whitegrid")
ax = sns.stripplot(x='class',y ='petal width', data=dsiris)
plt.title('Graph')
plt.show()
```

```python
ax = sns.swarmplot(x='class',y ='petal length', data=dsiris)
plt.title('Graph')
plt.show()
```

```python
ax = sns.swarmplot(x='petal length',y ='class', data=dsiris)
plt.title('Graph')
plt.show()
```
