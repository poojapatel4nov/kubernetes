```python
#Set Data column as timestamp and Name column as index of data.
df = pd.read_csv('hrdata.csv',index_col='Name',parse_dates=["HireDate"])
print(df)
```

```python
#change Header , need to setup header = 0 to change Header 
df = pd.read_csv('hrdata.csv',
                 index_col='Employee',
                 parse_dates=["Hire Date"],
                 header=0, names=["Employee","Hire Date","Salary", "Sick Date"])
print(df)
```
